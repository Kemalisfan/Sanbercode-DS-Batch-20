```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
# code here
fig, ax = plt.subplots(figsize=(20,8), nrows=3, ncols=1, sharey=True )

ax[0].plot(x,tan, label='tanx',color='b')
ax[0].set_title('Grafik tanx')

ax[1].plot(x,cos, label='sinx',color='r')
ax[1].set_title('Grafik cosx')

ax[2].plot(x,sin, label='cosx',color='g')
ax[2].set_title('Grafik sinx')

fig.legend(loc='upper center')

plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
# code here
fig,ax = plt.subplots(figsize=(16,10),sharex=True)
ax.plot(x,y, color='b')
ax.set_title('Grafik CosX')

upper_left = ax.inset_axes([-0.1,0.9,0.3,0.3])
upper_left.plot(x,y5, color='y')
upper_left.plot(x,y, color='b')
upper_left.set_title('Grafik CosX^5')

upper_right = ax.inset_axes([0.9,0.9,0.3,0.3])
upper_right.plot(x,y4, color='black')
upper_right.plot(x,y, color='b')
upper_right.set_title('Grafik CosX^4')

lower_left = ax.inset_axes([-0.1,-0.1,0.3,0.3])
lower_left.plot(x,y2, color='r')
lower_left.plot(x,y, color='b')
lower_left.set_title('Grafik CosX^2')

lower_right = ax.inset_axes([0.9,-0.1,0.3,0.3])
lower_right.plot(x,y3, color='g')
lower_right.plot(x,y, color='b')
lower_right.set_title('Grafik CosX^3')

plt.show()
```
