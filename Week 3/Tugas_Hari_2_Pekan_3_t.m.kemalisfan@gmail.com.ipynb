Dengan menggunakan data di bawah ini, plot kan data tersebut kemudian custom sehingga memiliki result seperti yang di harapkan :

note: warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
import numpy as np

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
# code here
import matplotlib.pyplot as plt

fig, ax=plt.subplots(figsize=(12,8))
ax.plot(x,cos, linestyle='-', color='Red', label='Cosx')
ax.plot(x,cos3, linestyle='dashed',color='Green',label='3 Cosx')
ax.plot(x,sin, linestyle='-.', color ='blue', label='Sin(x-90)')
ax.plot(x,sin2, linestyle='none',marker='x', color ='y', label='2 Sinx')
ax.legend(loc='center',bbox_to_anchor=(0.476, 0.483))
ax.set_xlabel('X-Axis Label')
ax.set_ylabel('Y-Axis Label')
ax.set_title('Grafik Cos Sin')
plt.show()
```
