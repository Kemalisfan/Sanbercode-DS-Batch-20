Pelajarilah secara garis besar suatu model/algoritma machine learning, kemudian aplikasikan untuk membuat model bagi data (variable x, y) di bawah ini. Kemudian buatlah prediksi terhadap data training dan data baru di interval 20-30.

- Plot data asli, data hasil prediksi terhadap data training, data hasil prediksi terhadap data baru
- Tunjukan beberapa parameter yang dimiliki model yang telah kalian buat

```python
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.RandomState(42)
x = 20 * rng.rand(50)
y = x**2 + 2 * x + - 1 + rng.randn(50)

# data yang akan di prediksi
x_new = np.arange(20, 30, 0.5)

fig, ax = plt.subplots(figsize=(12, 6))
ax.scatter(x, y)
ax.set_xlabel('X-Value')
ax.set_ylabel('Y-Value')
plt.show()
```

### Linear Regression

```python
# Code here
# 1. Memilih model algoritma 
from sklearn.linear_model import LinearRegression

x_train = x[:, np.newaxis]  #membuat variabel training x_train dengan dimensi = 2
y_train = y[:, np.newaxis]  #membuat variabel training y_train dengan dimensi = 2
x_test = x_new[:, np.newaxis] #membuat variabel x_new dengan dimensi = 2 untuk testing

# 2. Menentukan hyperparameters
reg = LinearRegression()

# 3. Melakukan training 
reg.fit(x_train, y_train)

# 4. lakukan prediksi dari model
y_train_pred = reg.predict(x_train)
y_pred = reg.predict(x_test)
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.scatter(x, y, label = 'Original Data')
ax.scatter(x_train, y_train_pred, label='Training Data Prediction')
ax.plot(x_new, y_pred, label = 'Predicted New Data' )
ax.set_xlabel('X-Value')
ax.set_ylabel('Y-Value')
plt.title('Linear Regression')
plt.legend()
plt.show()
```

### K Nearest Neighbors

```python
# 1. Memilih model algoritma 
from sklearn import neighbors

x_train = x[:, np.newaxis]  #membuat variabel training x_train dengan dimensi = 2
y_train = y[:, np.newaxis]  #membuat variabel training y_train dengan dimensi = 2
x_test = x_new[:, np.newaxis] #membuat variabel x_new dengan dimensi = 2 untuk testing

# 2. Menentukan hyperparameters
knn = neighbors.KNeighborsRegressor(weights='uniform', n_neighbors=5)

# 3. Melakukan training 
knn.fit(x_train,y_train)

# 4. lakukan prediksi dari model
y_train_pred2 = knn.predict(x_train) 
y_pred2 = knn.predict(x_test)
    
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.scatter(x, y, label = 'Original Data')
ax.scatter(x_train, y_train_pred2, label='Training Data Prediction')
ax.plot(x_new, y_pred2, label = 'Predicted New Data')
ax.set_xlabel('X-Value')
ax.set_ylabel('Y-Value')
plt.title('K Nearest Neighbors')
plt.legend()
plt.show()
```

Parameter used: 
                
                n_neighbors = 5
                
                weights = 'uniform

#### Random Forest Regressor

```python
# 1. pilih model algoritma
from sklearn.ensemble import RandomForestRegressor

x_train = x[:, np.newaxis]  #membuat variabel training x_train dengan dimensi = 2
y_train = y[:, np.newaxis]  #membuat variabel training y_train dengan dimensi = 2
x_test = x_new[:, np.newaxis] #membuat variabel x_new dengan dimensi = 2 untuk testing

# 2. tentukan hyperparameters
RFreg = RandomForestRegressor(n_estimators = 50, criterion='mae', max_depth=30, min_samples_split=2)

# 3. lakukan pembelajaran/training
RFreg.fit(x_train, y_train)

# 4. lakukan prediksi dari model
y_train_pred3 = RFreg.predict(x_train)
y_pred3 = RFreg.predict(x_test)
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.scatter(x, y, label='Data asli')
ax.scatter(x_train, y_train_pred3, label='Prediksi data training')
ax.plot(x_new, y_pred3, label='Prediksi data baru')
ax.set_xlabel('X-Value')
ax.set_ylabel('Y-Value')
plt.title('Random Forest Regression')
plt.legend()
plt.show()
```

Parameter used : 
                    
                    n_estimators      = 50   
                    criterion         = 'mae'
                    max_depth         = 30
                    min_samples_split = 2
